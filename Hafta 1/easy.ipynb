Kullanıcının doğum tarihini alarak kaç yaşında olduğunu bulan bir algoritma yazmanızı istiyorum.

```python
import datetime

def hesapla_yas(dogum_tarihi):
    simdiki_tarih = datetime.datetime.now()
    yas = simdiki_tarih.year - dogum_tarihi.year
    if (
        simdiki_tarih.month < dogum_tarihi.month
        or (
            simdiki_tarih.month == dogum_tarihi.month
            and simdiki_tarih.day < dogum_tarihi.day
              )
    ):
        yas -= 1
    return yas
dogum_tarihi = input("Doğum tarihinizi (GG/AA/YYYY) formatında girin: ")
dogum_tarihi = datetime.datetime.strptime(dogum_tarihi, "%d/%m/%Y")
yas = hesapla_yas(dogum_tarihi)
print("Yaşınız:", yas)
```
